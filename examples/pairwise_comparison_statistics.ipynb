```python
import pandas as pd
from decimal import Decimal
from scipy.stats import wilcoxon
```

```python
#convert numeric to scientific notation
def sci_notation(number, sig_fig=2):
    ret_string = "{0:.{1:d}e}".format(number, sig_fig)
    a, b = ret_string.split("e")
    # # remove leading "+" and strip leading zeros
    b = int(b)
    return a + " <10" + str(b)
```

```python
cpdp_results = pd.read_csv('https://raw.githubusercontent.com/jsaj/MDS_CPDP/master/benchmark-execution/benchmarks/CPDP_results.csv?token=GHSAT0AAAAAAB3VLK7U665Q6TJZUBAU7GK2Y5PGFKA')
```

```python
import warnings
warnings.filterwarnings("ignore")

result = []
for metric in ['fscore', 'auc', 'pf']:
 
  for data in ['AEEEM', 'NASA', 'PROMISE', 'RELINK']:
    mds_cpdp = cpdp_results.loc[(cpdp_results['Dataset'] == data) &
                           (cpdp_results['Method'] == 'MDS-CPDP')]
    values = []
    cols = []
    for methods in ['CamargoCruz09-DT', 'EASC-NB', 'Menzies11-RF', 'Turhan09-DT', 'Watanabe08-DT']:
      method_df = cpdp_results.loc[(cpdp_results['Dataset'] == data) &
                                  (cpdp_results['Method'] == methods)]
                                   
      cols.append(methods)
      mds_values = mds_cpdp[metric]
      method_df = method_df[metric]
      s, p = wilcoxon(mds_values, method_df)
      values.append(sci_notation(p, sig_fig=4))

    df = pd.DataFrame([values], columns=cols)

    df.insert(0, 'Datasets', data)
    df.insert(0, 'Metric', metric)
    result.append(df)

result = pd.concat(result).reset_index(drop=True)
result
```
