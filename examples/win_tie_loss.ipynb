# Imports

```python
import numpy as np
import pandas as pd
```

# Read data

```python
df = pd.read_csv('/content/Multi_DES/benchmark-execution/benchmarks/CPDP_results.csv')
df.head()
```

# Count win-tie-loss by metric vs dataset

```python
results = []
for metric in ['fscore', 'auc', 'pf']:
  for data in ['AEEEM', 'NASA', 'PROMISE', 'RELINK']:
    res = []
    for m in ['CamargoCruz09-DT', 'EASC-NB', 'Menzies11-RF', 'Turhan09-DT', 'Watanabe08-DT']:
      win = 0
      tie = 0
      loss = 0
      df_data = df.loc[df['Dataset'] == data]
      for project in np.unique(df_data['ProductName']):
        df_project = df_data.loc[df_data['ProductName'] == project]
        
    

        df_metric = df_project[['Method', metric]].reset_index(drop=True).round(3)
        multi_score = df_metric.loc[df_metric['Method'] == 'Multi-DES', metric].values[0]
        if metric != 'pf': 
          cpdp_score = df_metric.loc[df_metric['Method'] == m, metric].values[0]
          if multi_score < cpdp_score:
            loss += 1
          elif  multi_score == cpdp_score:
            tie += 1
          else:
            win +=1

        else:
          cpdp_score = df_metric.loc[df_metric['Method'] == m, metric].values[0]
          if multi_score > cpdp_score:
            loss += 1
          elif  multi_score == cpdp_score:
            tie += 1
          else:
            win +=1

      values = ['{}-{}-{}'.format(win, tie, loss)]
      res.append(pd.DataFrame([values], columns=[m]))
    res = pd.concat(res, axis=1)
    res.insert(0, 'Datasets', data)
    res.insert(0, 'Metric', metric)
    results.append(res)
results = pd.concat(results).reset_index(drop=True)
```
