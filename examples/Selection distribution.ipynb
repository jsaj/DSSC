```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

results_df = pd.read_csv("configs_mds_cpdp.csv")
```

# Renaming dataframe to show proper method names used in the paper

```python
results_df['dynamicSelection'][results_df['dynamicSelection']=='METADES'] = 'META-DES'
results_df['dynamicSelection'][results_df['dynamicSelection']=='KNORAE'] = 'KNE'
results_df['dynamicSelection'][results_df['dynamicSelection']=='KNORAU'] = 'KNU'
results_df['classifier'][results_df['classifier']=='RandomForestClassifier'] = 'Random Forest'
results_df['classifier'][results_df['classifier']=='LogisticRegression'] = 'Logistic Regression'
results_df['classifier'][results_df['classifier']=='DecisionTreeClassifier'] = 'Decision Tree'
results_df['classifier'][results_df['classifier']=='GaussianNB'] = 'Naive Bayes'
```

```python
results_df.columns
ax = results_df['dynamicSelection'].value_counts(dropna=False).plot.bar()
#ax.set_xlabel("Dynamic Selection technique", fontsize=16)
ax.set_ylabel("No. Selection", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.savefig('SelectionDSMethods.pdf', dpi=600)
```

```python
results_df['dynamicSelection'].value_counts(dropna=False)
```

```python
ax = results_df['classifier'].value_counts(dropna=False).plot.bar()
#ax.set_xlabel("Base classifier model", fontsize=16)
ax.set_ylabel("No. Selection", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.savefig('SelectionBaseClassifier.pdf', dpi=600)
```

```python
results_df['classifier'].value_counts(dropna=False)
```

```python
ax = results_df['n_estimators'].value_counts(dropna=False).sort_index().plot.bar()
#ax.set_xlabel("Pool size", fontsize=16)
ax.set_ylabel("No. Selection", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.savefig('SelectionNEstimators.pdf', dpi=600)
```
